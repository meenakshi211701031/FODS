```python
q1=df['OrderPrice'].quantile(0.25)
```

```python
q1
```

```python
q3=df['OrderPrice'].quantile(0.75)
```

```python
q3
```

```python
IQR=q3-q1
```

```python
lb=q1-1.5*IQR
```

```python
ub=q3+1.5*IQR
```

```python
outlier=((df[['OrderPrice']]<lb) | (df[['OrderPrice']]>ub)).any(axis=1)
```

```python
cleaned=df[~outlier]
```
