```python
df=pd.read_csv('inventory_dataset - inventory_dataset.csv')
```

```python
null=df.columns[df.isnull().any()]
```

```python
median=df['Purchase_Price'].median()
```

```python
df['Purchase_Price'].fillna(median,inplace=True)
```

```python
df
```

```python
mean = df['Sale_Price'].mean()
```

```python
df['Sale_Price'].fillna(mean,inplace=True)
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
Q1 = df['Sale_Price'].quantile(0.25)
```

```python
Q3 = df['Sale_Price'].quantile(0.75)
```

```python
IQR=Q3-Q1
```

```python
lb=Q1 - 1.5 * IQR
```

```python
lb
```

```python
up=Q3+1.5 * IQR
```

```python
up
```

```python
Sale_price=df[(df['Sale_Price'] >=lb) & (df['Sale_Price'] <=up)]
```

```python
Sale_price
```

```python
scaler = MinMaxScaler()
```

```python
scaler
```

```python
min_quan=df['Quantity_In_Stock'].min()
```

```python
max_quan=df['Quantity_In_Stock'].max()
```

```python
df['Normalize_Quantity']=(df['Quantity_In_Stock']-min_quan /(max_quan-min_quan))
```

```python
df
```

```python
df=df.drop('Normalize_Quantity_In_Stock',axis=1)
```

```python
df
```

```python
df=df.drop('Normalize_Quantity',axis=1)
```

```python
df
```

```python
df['Quantity_In_Stock']= scaler.fit_transform(df[['Quantity_In_Stock']])
```

```python
df
```

```python
df['Quantity_In_Stock'] = df['Quantity_In_Stock'].astype(int)
```

```python
df
```

```python
df['Discontinued'] = df['Discontinued'].map({'Yes': 1, 'No': 0})
```

```python
df
```

```python
df.to_csv('cleaned_dataset.csv', index=False)
```
